```python
%matplotlib inline 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import requests
response = requests.get('http://seanlahman.com/files/database/lahman-csv_2014-02-14.zip')

from io import BytesIO
from zipfile import ZipFile
from urllib.request import urlopen
resp = urlopen('http://seanlahman.com/files/database/lahman-csv_2014-02-14.zip')
with ZipFile(BytesIO(resp.read())) as thezip:
    with thezip.open("Salaries.csv") as salaries:
        salarydf = pd.read_csv(salaries)
    with thezip.open("Teams.csv") as teams:
        teamsdf = pd.read_csv(teams)
```

```python
totals = pd.DataFrame(salarydf.groupby(['teamID', 'yearID'])['salary'].sum())
merged = pd.merge(totals, teamsdf, on=['teamID', 'yearID'])
wbs = merged[['teamID','yearID','W','salary']] # wbs = wins by salary
```

```python
result = {each_year : pd.DataFrame(wbs[wbs.yearID == each_year]) for each_year in wbs.yearID.unique()}

# This will generate a separate scatter plot for every year (29 graphs)

# for each_year in result:
#     this_year = result[each_year]
#     this_year['color'] = 'blue'
#     idx = this_year.loc[this_year.teamID =='OAK'].index
#     this_year.loc[idx,'color'] = 'red'
#     this_year.plot.scatter('salary','W',c='color',title=this_year.yearID.values[0])

#Instead, let's make one graph with two lines, showing Oakland's wins and salary relative to other teams over time

oakland = pd.DataFrame(wbs[wbs.teamID == 'OAK'])
years = sorted(oakland.yearID.values)
win_position = []
salary_position = []
for each_year in years:
    wins_this_year = pd.DataFrame(wbs[wbs.yearID == each_year][['W', 'teamID']])
    wins_this_year = wins_this_year.sort_values(by='W', ascending = False)
    win_position.append(
        np.where(wins_this_year.teamID.values == 'OAK')[0][0]+1
    )
    salaries_this_year = pd.DataFrame(wbs[wbs.yearID == each_year][['salary', 'teamID']])
    salaries_this_year = salaries_this_year.sort_values(by='salary', ascending = False)
    salary_position.append(
        np.where(salaries_this_year.teamID.values == 'OAK')[0][0]+1
    )

win_position = pd.Series(index = years, data = win_position)
salary_position = pd.Series(index = years, data = salary_position)
year_index = pd.Index(years)
new = pd.DataFrame(index=year_index)
new['wins'] = win_position
new['salary'] = salary_position
new.plot(ylabel="League position", title="Oakland Athletics").invert_yaxis()
```
